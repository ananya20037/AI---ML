Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

```python
# Write your code from here
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

if 'Age' in df.columns:
    plt.hist(df['Age'].dropna(), bins=20, edgecolor='black', color='skyblue')
    plt.title('Histogram of Age')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("No 'Age' feature found in the dataset.")
```

Title: Correlation Heatmap<br>

Question 5 : Generate a correlation heatmap for a dataset with multiple features (e.g., height ,weight , age ) and explain the correlations observed.

```python
# Write your code from here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {'Height': [150, 160, 170, 175, 180, 185, 190, 160, 170, 155, 165, 180, 185, 175, 165],
        'Weight': [50, 60, 70, 75, 80, 85, 90, 60, 70, 55, 65, 80, 85, 75, 65],
        'Age': [25, 30, 35, 40, 45, 50, 55, 30, 35, 28, 33, 40, 45, 38, 34]}
df = pd.DataFrame(data)

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python
# Write your code from here
import pandas as pd
from scipy.stats import skew, kurtosis


data = {'Temperature': [36.5, 36.7, 36.8, 37.0, 37.1, 36.6, 36.9, 37.2, 36.8, 36.7, 37.0, 37.1, 36.6, 36.5, 36.9]}
df = pd.DataFrame(data)

mean_val = df['Temperature'].mean()
median_val = df['Temperature'].median()
std_val = df['Temperature'].std()
skew_val = skew(df['Temperature'])
kurt_val = kurtosis(df['Temperature'])

print(f'Mean: {mean_val:.2f}')
print(f'Median: {median_val:.2f}')
print(f'Standard Deviation: {std_val:.2f}')
print(f'Skewness: {skew_val:.2f}')
print(f'Kurtosis: {kurt_val:.2f}')
```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
# Write your code from here
import pandas as pd

data = {'Price': [10, 12, 14, 16, 18, 20, 22],
        'Demand': [100, 95, 90, 85, 80, 75, 70]}
df = pd.DataFrame(data)

covariance = df['Price'].cov(df['Demand'])
correlation = df['Price'].corr(df['Demand'])

print(f'Covariance: {covariance:.2f}')
print(f'Correlation: {correlation:.2f}')
```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python
# Write your code from here
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


data = {
    'x1': [2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2.0, 1.0, 1.5, 1.1],
    'x2': [2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9],
    'x3': [1.2, 0.9, 2.5, 2.1, 2.6, 2.4, 1.7, 1.0, 1.8, 1.1],
    'x4': [1.9, 1.2, 2.8, 2.5, 3.3, 2.5, 2.0, 1.4, 2.0, 1.3]
}
df = pd.DataFrame(data)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

plt.scatter(pca_df['PC1'], pca_df['PC2'], color='green')
plt.title('PCA - 2 Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```
