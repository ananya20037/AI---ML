Title: Correlation Heatmaps<br>

Task 1: Generate a correlation heatmap of the dataset.<br>
Task 2: Highlight correlation between sepal length and petal length.<br>
Task 3: Highlight correlation between petal width and petal length.<br>

```python
# Write your code from hereimport pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

plt.figure(figsize=(8, 6))
mask_1 = correlation_matrix[['sepal length (cm)', 'petal length (cm)']].notna()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', mask=~mask_1)
plt.title('Highlighted Correlation between Sepal Length and Petal Length')
plt.show()

plt.figure(figsize=(8, 6))
mask_2 = correlation_matrix[['petal width (cm)', 'petal length (cm)']].notna()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', mask=~mask_2)
plt.title('Highlighted Correlation between Petal Width and Petal Length')
plt.show()
```

**Multivariate Analysis (Higher-Dimensional Data Relationships)**<br>
Objective: Analyze relationships in higher-dimensional data using pair plots and PCA.

Title: Pair Plots<br>

Task 1: Create a pair plot for the Iris dataset.<br>
Task 2: Focus on a subset of features (e.g., only petal dimensions).<br>
Task 3: Exclude one class to observe differences.

```python
# Write your code from here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = data.target

sns.pairplot(df, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()

df_petal = df[['petal length (cm)', 'petal width (cm)', 'species']]
sns.pairplot(df_petal, hue='species')
plt.title('Pair Plot for Petal Dimensions')
plt.show()

df_no_class_0 = df[df['species'] != 0]
sns.pairplot(df_no_class_0, hue='species')
plt.title('Pair Plot Excluding Class 0 (Setosa)')
plt.show()
```

Title: Principal Component Analysis (PCA)<br>

Task 1: Perform PCA and plot the first two principal components.<br>
Task 2: Visualize explained variance by each component.<br>
Task 3: Retain more components and visualize in 3D.

```python
# Write your code from here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = data.target

pca = PCA(n_components=2)
pca_result = pca.fit_transform(df.drop('species', axis=1))

plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=df['species'], palette='viridis')
plt.title('PCA - First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(8, 6))
sns.barplot(x=[1, 2], y=explained_variance)
plt.title('Explained Variance by Each Principal Component')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance')
plt.show()

pca_3d = PCA(n_components=3)
pca_result_3d = pca_3d.fit_transform(df.drop('species', axis=1))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_result_3d[:, 0], pca_result_3d[:, 1], pca_result_3d[:, 2], c=df['species'], cmap='viridis')
ax.set_title('PCA - 3D Visualization')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
```

**Statistical Analysis in EDA**<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
# Write your code from here
import pandas as pd
from scipy.stats import skew, kurtosis
from sklearn.datasets import load_iris

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

mean_petal_length = df['petal length (cm)'].mean()
median_petal_length = df['petal length (cm)'].median()
std_petal_length = df['petal length (cm)'].std()

skew_sepal_width = skew(df['sepal width (cm)'])
kurt_sepal_width = kurtosis(df['sepal width (cm)'])

mean_sepal_length = df['sepal length (cm)'].mean()
median_sepal_length = df['sepal length (cm)'].median()
std_sepal_length = df['sepal length (cm)'].std()

mean_petal_length, median_petal_length, std_petal_length, skew_sepal_width, kurt_sepal_width, mean_sepal_length, median_sepal_length, std_sepal_length
```

Title: Correlation & Covariance<br>

Task 1: Compute correlation between sepal length and petal length.<br>
Task 2: Compute covariance between petal width and sepal width.<br>
Task 3: Determine the most correlated pair of features.<br>

```python
# Write your code from here
import pandas as pd
from sklearn.datasets import load_iris

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

correlation_sepal_petal = df['sepal length (cm)'].corr(df['petal length (cm)'])

covariance_petal_sepal = df['petal width (cm)'].cov(df['sepal width (cm)'])

correlation_matrix = df.corr()
most_correlated_pair = correlation_matrix.unstack().idxmax()
most_correlated_value = correlation_matrix.unstack().max()

correlation_sepal_petal, covariance_petal_sepal, most_correlated_pair, most_correlated_value
```
