Question 8: Correlation Analysis
<br>
Description: Calculate and visualize the correlation matrix for numerical features.

```python
# Write your code from here
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/workspaces/AI---ML/src/Module 2/Hands-on Scikit-Learn Machine Learning Models Data Splitting and Data Aggregation using Python/forest_health_data.csv')

# Select only numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

Question 10: Advanced Outlier Detection
<br>
Description: Use the Z-score method to identify and handle outliers.

```python
# Write your code from here
import numpy as np
import pandas as pd
from scipy import stats

df = pd.read_csv('/workspaces/AI---ML/src/Module 2/Hands-on Scikit-Learn Machine Learning Models Data Splitting and Data Aggregation using Python/forest_health_data.csv')

# Calculate the Z-scores for numerical columns
numeric_df = df.select_dtypes(include=['number'])
z_scores = np.abs(stats.zscore(numeric_df))

# Set a threshold for identifying outliers (commonly 3)
threshold = 3

# Identify outliers
outliers = (z_scores > threshold).all(axis=1)

# Filter out the outliers
df_no_outliers = df[~outliers]

print(df_no_outliers.head())
```
