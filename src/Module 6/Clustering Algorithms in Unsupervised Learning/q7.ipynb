```python
# Question 6 : Comparing K-Means and Hierarchical Clustering
# Description: Compare the results of K-Means and Hierarchical Clustering on a dataset and discuss the potential differences in their clustering patterns.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=3, n_features=2, random_state=42)

kmeans_labels = KMeans(n_clusters=3, random_state=42).fit_predict(X)
hierarchical_labels = AgglomerativeClustering(n_clusters=3).fit_predict(X)

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('K-Means Clustering')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=hierarchical_labels, cmap='viridis')
plt.title('Hierarchical Clustering')

plt.show()
```
