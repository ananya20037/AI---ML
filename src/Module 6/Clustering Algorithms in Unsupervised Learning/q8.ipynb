```python
# Question 8: Parameter Tuning in DBSCAN
# Description: Experiment with different values of eps and min_samples in DBSCAN and observe the effect on clustering results.
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

params = [(0.1, 5), (0.2, 5), (0.3, 5), (0.2, 3), (0.2, 10)]

plt.figure(figsize=(15, 10))
for i, (eps, min_samples) in enumerate(params):
    model = DBSCAN(eps=eps, min_samples=min_samples)
    labels = model.fit_predict(X)
    plt.subplot(2, 3, i + 1)
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
    plt.title(f'eps={eps}, min_samples={min_samples}')
plt.tight_layout()
plt.show()
```
