```python
# Question 3: Implement t-SNE on a Dataset
# Description: Use t-SNE to visualize high-dimensional data.
from sklearn.manifold import TSNE
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, n_features=10, random_state=42)

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c='blue', edgecolor='k', s=50)
plt.title('t-SNE: 2D Visualization of High-Dimensional Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
