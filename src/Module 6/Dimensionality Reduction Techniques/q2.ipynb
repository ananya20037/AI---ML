```python
# Question 2: Geometry of PCA
# Description: Compute PCA on a simple 2D matrix and manually derive its principal components.
import numpy as np
import matplotlib.pyplot as plt

# Sample 2D dataset
X = np.array([[2, 3], [3, 4], [5, 6], [7, 8], [8, 9]])

# Step 1: Center the data (subtract mean)
mean = np.mean(X, axis=0)
X_centered = X - mean

# Step 2: Compute the covariance matrix
cov_matrix = np.cov(X_centered.T)

# Step 3: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sort eigenvalues and eigenvectors
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]

# Step 5: Project data onto the principal components
X_pca = X_centered.dot(eigenvectors)

# Plotting original data and principal components
plt.scatter(X[:, 0], X[:, 1], color='blue', label='Original Data')
plt.quiver(mean[0], mean[1], eigenvectors[0, 0], eigenvectors[1, 0], angles='xy', scale_units='xy', scale=1, color='red', label='PC1')
plt.quiver(mean[0], mean[1], eigenvectors[0, 1], eigenvectors[1, 1], angles='xy', scale_units='xy', scale=1, color='green', label='PC2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('PCA: Geometry and Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```
