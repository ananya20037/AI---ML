```python
#Title: Perform K-Means Clustering on a Dataset

# Task 1: Clustering Customer Data
# Objective: Cluster a dataset of customer purchases to identify customer segments.
# Dataset: Use a publicly available dataset like the "Mall Customer Segmentation Dataset."

# Task 2: Clustering Plant Species
# Objective: Group different plant species based on their attributes.
# Dataset: Use the "Iris" dataset from Scikit-learn.

# Task 3: Clustering for Image Compression
# Objective: Use K-Means clustering for image compression.
# Dataset: Choose a sample image from Scikit-learns datasets.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_sample_image
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Task 1: Clustering Customer Data (Using artificial data)

# Simulating customer data for demonstration (Annual Income and Spending Score)
np.random.seed(42)
annual_income = np.random.randint(30, 150, size=200)  # Annual income in k$
spending_score = np.random.randint(1, 101, size=200)  # Spending score from 1 to 100

customer_data = np.column_stack((annual_income, spending_score))

# Apply K-Means clustering
kmeans_customer = KMeans(n_clusters=5, random_state=42)
customer_clusters = kmeans_customer.fit_predict(customer_data)

# Plotting Customer Data Clusters
plt.figure(figsize=(8, 6))
plt.scatter(customer_data[:, 0], customer_data[:, 1], c=customer_clusters, cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation')
plt.colorbar()
plt.show()


# Task 2: Clustering Plant Species (Using Iris dataset)

# Load the iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target

# Apply K-Means clustering
kmeans_iris = KMeans(n_clusters=3, random_state=42)
iris_clusters = kmeans_iris.fit_predict(X_iris)

# Plotting Iris Data Clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=iris_clusters, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset Clustering')
plt.colorbar()
plt.show()


# Task 3: Clustering for Image Compression (Using a Sample Image)

# Load a sample image from sklearn datasets
china = load_sample_image("china.jpg")

# Reshape the image to a 2D array (each pixel as a row)
china_resized = china.reshape(-1, 3)

# Apply K-Means for image compression (reduce to 16 colors)
kmeans_image = KMeans(n_clusters=16, random_state=42)
image_compressed = kmeans_image.fit_predict(china_resized)
china_compressed_image = kmeans_image.cluster_centers_[image_compressed].reshape(china.shape)

# Plot original and compressed images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(china)
axes[0].set_title('Original Image')
axes[0].axis('off')
axes[1].imshow(china_compressed_image)
axes[1].set_title('Compressed Image (16 colors)')
axes[1].axis('off')
plt.show()
```
