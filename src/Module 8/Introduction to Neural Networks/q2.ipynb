```python
# Question 2: Compare Biological and Artificial Neurons using Code
# Description: Visualize how biological neurons and artificial neurons process information differently.

import numpy as np
import matplotlib.pyplot as plt

# Biological Neuron Model (Simplified)
def biological_neuron(input_signal):
    # Biological neurons are more complex, but we can model them simply as passing through an activation function.
    # Here, we'll use a sigmoid function to represent a simple biological neuron.
    output_signal = 1 / (1 + np.exp(-input_signal))  # Sigmoid Activation
    return output_signal

# Artificial Neuron Model
def artificial_neuron(inputs, weights, bias):
    # Sum of weighted inputs + bias
    weighted_sum = np.dot(inputs, weights) + bias
    # ReLU Activation (simplified for an artificial neuron)
    output_signal = np.maximum(0, weighted_sum)  # ReLU Activation
    return output_signal

# Visualization Setup
inputs = np.array([0.5, 0.8])  # Example inputs to both neurons
weights = np.array([0.6, 0.9])  # Weights for the artificial neuron
bias = 0.1  # Bias for the artificial neuron

# Biological Neuron: Apply Sigmoid Activation
biological_output = biological_neuron(np.sum(inputs))  # Biological neuron just sums inputs

# Artificial Neuron: Apply Weighted Sum + Bias followed by ReLU
artificial_output = artificial_neuron(inputs, weights, bias)

# Print Results
print(f"Biological Neuron Output (Sigmoid): {biological_output}")
print(f"Artificial Neuron Output (ReLU): {artificial_output}")

# Plotting the difference in outputs for visualization
x = np.linspace(-5, 5, 100)
biological_y = 1 / (1 + np.exp(-x))  # Sigmoid curve for biological neuron
artificial_y = np.maximum(0, x)  # ReLU curve for artificial neuron

plt.figure(figsize=(10, 6))

# Plot Sigmoid Function for Biological Neuron
plt.subplot(1, 2, 1)
plt.plot(x, biological_y, label="Sigmoid (Biological Neuron)", color='b')
plt.title('Biological Neuron (Sigmoid Activation)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Plot ReLU Function for Artificial Neuron
plt.subplot(1, 2, 2)
plt.plot(x, artificial_y, label="ReLU (Artificial Neuron)", color='r')
plt.title('Artificial Neuron (ReLU Activation)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
