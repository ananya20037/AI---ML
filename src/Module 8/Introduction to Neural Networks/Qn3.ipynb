```python
# Biological VS Artificial Neural Networks

# Task 1: Human Brain vs Digital Neural Network
# Task: Compare the structure and functionality of a biological neural network with that of an artificial one.
# Exercise: List three similarities and three differences between them.

# Task 2: Signal Processing
# Task: Understand how signals are processed in both biological neurons and artificial neurons.
# Exercise: Draw a flowchart of how information travels through both types of networks.

# Task 3: Learning Mechanisms
# Task: Examine the learning process in natural and artificial networks.
# Exercise: Create a table comparing synaptic plasticity with weight adjustments in artificial networks.


# --- Task 1: Human Brain vs Digital Neural Network ---

similarities = [
    "Both have interconnected units (neurons or nodes) that process information.",
    "Both adapt and learn from input stimuli (experience/data).",
    "Both use signal transmission to influence subsequent layers or neurons."
]

differences = [
    "Biological neurons use electrochemical signals; artificial neurons use mathematical functions.",
    "Biological networks are massively parallel and energy-efficient; artificial networks require significant computation.",
    "Learning in the brain involves synaptic plasticity; in ANNs, it involves numerical weight updates using algorithms like backpropagation."
]

print("Similarities:")
for s in similarities:
    print("-", s)

print("\nDifferences:")
for d in differences:
    print("-", d)


# --- Task 2: Signal Processing Flowcharts ---

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def draw_flowchart(title, steps):
    fig, ax = plt.subplots(figsize=(5, len(steps)))
    ax.set_xlim(0, 1)
    ax.set_ylim(0, len(steps))
    ax.axis('off')
    for i, step in enumerate(steps):
        ax.add_patch(mpatches.FancyBboxPatch((0.1, len(steps)-i-1+0.2), 0.8, 0.6, boxstyle="round,pad=0.1", facecolor="#a1d8f0"))
        ax.text(0.5, len(steps)-i-1+0.5, step, ha="center", va="center", fontsize=10)
        if i < len(steps)-1:
            ax.annotate('', xy=(0.5, len(steps)-i-1), xytext=(0.5, len(steps)-i-0.4), arrowprops=dict(arrowstyle="->"))
    plt.title(title, fontsize=12)
    plt.show()

bio_steps = [
    "Stimulus received by sensory neuron",
    "Electrical impulse travels along axon",
    "Neurotransmitters released at synapse",
    "Signal received by next neuron",
    "Response triggered"
]

ann_steps = [
    "Input data fed to input layer",
    "Weighted sum and activation function",
    "Forward pass through hidden layers",
    "Output layer produces prediction",
    "Loss calculated and backpropagated"
]

draw_flowchart("Biological Neural Network Signal Flow", bio_steps)
draw_flowchart("Artificial Neural Network Signal Flow", ann_steps)


# --- Task 3: Learning Mechanisms Comparison ---

import pandas as pd

data = {
    "Aspect": ["Mechanism", "Basis of Change", "Learning Process", "Timescale", "Error Handling", "Flexibility"],
    "Biological Neural Networks": [
        "Synaptic Plasticity",
        "Strengthening/weakening of synapses",
        "Hebbian learning / long-term potentiation",
        "Milliseconds to years",
        "Redundant and distributed, not error-driven",
        "High flexibility with lifelong learning"
    ],
    "Artificial Neural Networks": [
        "Weight Adjustment",
        "Change in weights via gradients",
        "Backpropagation with gradient descent",
        "Milliseconds to minutes (training cycles)",
        "Error-driven (loss minimization)",
        "Limited flexibility, needs retraining"
    ]
}

df = pd.DataFrame(data)
print(df.to_string(index=False))
```
