Sub-Title: Confusion Matrix

Task 1: Construct and interpret a confusion matrix for tumor classification.

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target  # 0: malignant, 1: benign

if X.empty or y.size == 0:
    raise ValueError("Input features or target is empty.")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

if X_train.empty or y_train.size == 0:
    raise ValueError("Training data is empty.")

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

if len(y_pred) != len(y_test):
    raise ValueError("Mismatch in prediction and ground truth sizes.")

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Malignant', 'Benign'])
disp.plot(cmap='Blues')
print(f"Confusion Matrix:\n{cm}")
```

Task 2: Create a confusion matrix for fraud detection model evaluation.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5, 
                           random_state=42, weights=[0.9, 0.1])

if X.size == 0 or y.size == 0:
    raise ValueError("Input data is empty.")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

if X_train.size == 0 or y_train.size == 0:
    raise ValueError("Training set is empty.")

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

if len(y_pred) != len(y_test):
    raise ValueError("Prediction count does not match ground truth count.")

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-Fraud', 'Fraud'])
disp.plot(cmap='Blues')
print(f"Confusion Matrix:\n{cm}")
```


Task 3: Use confusion matrix to evaluate model predicting employee retention.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=5, n_informative=3, n_redundant=2, 
                           random_state=42, weights=[0.8, 0.2])

if X.size == 0 or y.size == 0:
    raise ValueError("Dataset is empty.")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

if X_train.size == 0 or y_train.size == 0:
    raise ValueError("Training data is empty.")

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

if len(y_pred) != len(y_test):
    raise ValueError("Prediction and actual label size mismatch.")

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Left', 'Stayed'])
disp.plot(cmap='Blues')
print(f"Confusion Matrix:\n{cm}")


```
