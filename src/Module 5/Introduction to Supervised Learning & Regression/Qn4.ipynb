Polynomial Regression

Task 1: Apply polynomial regression to a dataset showing the relationship between age and income.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# Simulated data: Age vs Income
age = np.array([18, 22, 25, 30, 35, 40, 45, 50, 55, 60]).reshape(-1, 1)
income = np.array([20000, 25000, 28000, 35000, 40000, 45000, 48000, 50000, 52000, 55000])

# Polynomial feature transformation
poly = PolynomialFeatures(degree=3)
age_poly = poly.fit_transform(age)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(age_poly, income, test_size=0.2)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

# Plotting the results
plt.scatter(age, income, color='blue', label='Actual data')
plt.plot(age, model.predict(poly.transform(age)), color='red', label='Polynomial fit (degree=3)')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()

print("Polynomial Regression Predictions (Income):")
print(predictions)
# Write your code here
```

Task 2: Explore and visualize a higher-degree polynomial fitting compared to a simple linear regression.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split

# Simulated data: Age vs Income
age = np.array([18, 22, 25, 30, 35, 40, 45, 50, 55, 60]).reshape(-1, 1)
income = np.array([20000, 25000, 28000, 35000, 40000, 45000, 48000, 50000, 52000, 55000])

# Polynomial feature transformation for higher degrees
poly_2 = PolynomialFeatures(degree=2)
poly_3 = PolynomialFeatures(degree=3)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(age, income, test_size=0.2)

# Simple Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Polynomial Regression (degree 2 and degree 3)
poly_model_2 = LinearRegression()
X_train_poly_2 = poly_2.fit_transform(X_train)
poly_model_2.fit(X_train_poly_2, y_train)

poly_model_3 = LinearRegression()
X_train_poly_3 = poly_3.fit_transform(X_train)
poly_model_3.fit(X_train_poly_3, y_train)

# Predict
y_pred_linear = linear_model.predict(X_test)
y_pred_poly_2 = poly_model_2.predict(poly_2.transform(X_test))
y_pred_poly_3 = poly_model_3.predict(poly_3.transform(X_test))

# Visualization
plt.scatter(age, income, color='blue', label='Actual data')
plt.plot(age, linear_model.predict(age), color='green', label='Linear Regression')
plt.plot(age, poly_model_2.predict(poly_2.transform(age)), color='red', label='Polynomial Regression (degree=2)')
plt.plot(age, poly_model_3.predict(poly_3.transform(age)), color='orange', label='Polynomial Regression (degree=3)')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()

print("Linear Regression Predictions (Income):")
print(y_pred_linear)

print("\nPolynomial Regression Predictions (degree 2):")
print(y_pred_poly_2)

print("\nPolynomial Regression Predictions (degree 3):")
print(y_pred_poly_3)
```

Task 3: Dataset with cyclical patterns such as tides or circadian cycles.

```python
# Write your code here
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Simulating a circadian cycle (24-hour) dataset
hours = np.linspace(0, 24, 100)  # 100 time points over a 24-hour period
temperature = 10 + 5 * np.sin(2 * np.pi * hours / 24) + np.random.normal(0, 0.5, 100)  # Sinusoidal pattern with noise

# Applying sine and cosine transformations to encode cyclical features
X = np.column_stack([np.sin(2 * np.pi * hours / 24), np.cos(2 * np.pi * hours / 24)])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, temperature, test_size=0.2, random_state=42)

# Model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)

# Visualization
plt.scatter(hours, temperature, color='blue', label='Actual Temperature')
plt.plot(hours, model.predict(np.column_stack([np.sin(2 * np.pi * hours / 24), np.cos(2 * np.pi * hours / 24)])), color='red', label='Fitted Curve')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature')
plt.legend()
plt.show()

print("Predictions for Temperature (based on circadian cycle):")
print(predictions[:5])
```
