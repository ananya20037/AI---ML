```python
# Question 5: Experiment with DBSCAN Parameters (eps and min_samples)
# Description: Explore how changing DBSCAN parameters affects clustering.
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a simple 2D dataset
X = np.array([[1, 2],
              [2, 3],
              [2, 4],
              [8, 7],
              [8, 8],
              [25, 80]])

# Step 2: Experiment with different values of eps and min_samples

# Case 1: eps=3, min_samples=2
dbscan1 = DBSCAN(eps=3, min_samples=2)
y_dbscan1 = dbscan1.fit_predict(X)

# Case 2: eps=2, min_samples=2
dbscan2 = DBSCAN(eps=2, min_samples=2)
y_dbscan2 = dbscan2.fit_predict(X)

# Case 3: eps=3, min_samples=3
dbscan3 = DBSCAN(eps=3, min_samples=3)
y_dbscan3 = dbscan3.fit_predict(X)

# Step 3: Visualize the results

plt.figure(figsize=(15, 5))

# Plot for Case 1
plt.subplot(131)
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan1, cmap='viridis', marker='o')
plt.title('eps=3, min_samples=2')

# Plot for Case 2
plt.subplot(132)
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan2, cmap='viridis', marker='o')
plt.title('eps=2, min_samples=2')

# Plot for Case 3
plt.subplot(133)
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan3, cmap='viridis', marker='o')
plt.title('eps=3, min_samples=3')

plt.show()
```
