```python
# Question 4: Introduction to DBSCAN Clustering for Anomaly Detection
# Description: Apply DBSCAN on a dataset to identify core samples and anomalies.
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a simple 2D dataset
X = np.array([[1, 2],
              [2, 3],
              [2, 4],
              [8, 7],
              [8, 8],
              [25, 80]])

# Step 2: Apply DBSCAN with epsilon=3 and min_samples=2
dbscan = DBSCAN(eps=3, min_samples=2)
y_dbscan = dbscan.fit_predict(X)

# Step 3: Visualize the clusters and anomalies
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', marker='o')

plt.title('DBSCAN Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
