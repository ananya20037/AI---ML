```python
# Question 3: Hierarchical Clustering with a Different Linkage Criteria
# Description: Apply hierarchical clustering using different linkage criteria (average linkage).
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Generate a simple 2D dataset
X = np.array([[5, 7],
              [1, 2],
              [2, 4],
              [6, 8],
              [9, 10]])

# Step 2: Perform Agglomerative Hierarchical Clustering using 'average' linkage
Z = linkage(X, method='average')

# Step 3: Visualize the dendrogram
plt.figure(figsize=(8, 6))
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering (Average Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
