```python
# Question 2: Understanding Dendrograms
# Description: Generate a dendrogram to visualize the clustering process.
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Generate a simple 2D dataset
X = np.array([[5, 7],
              [1, 2],
              [2, 4],
              [6, 8],
              [9, 10]])

# Step 2: Perform Agglomerative Hierarchical Clustering using 'single' linkage
Z = linkage(X, method='single')

# Step 3: Visualize the dendrogram
plt.figure(figsize=(8, 6))
dendrogram(Z)
plt.title('Dendrogram for Hierarchical Clustering (Single Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
