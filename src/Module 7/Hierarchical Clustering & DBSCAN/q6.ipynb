```python
# Question 6: Construct a Custom Dendrogram for Better Insights
# Description: Build a custom dendrogram with annotated thresholds to determine optimal cluster number.
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Generate a simple 2D dataset
X = np.array([[1, 2],
              [2, 3],
              [3, 4],
              [8, 7],
              [8, 8],
              [25, 80]])

# Step 2: Perform hierarchical clustering
Z = linkage(X, method='ward')

# Step 3: Plot dendrogram with custom thresholds
plt.figure(figsize=(10, 6))
dendrogram(Z, color_threshold=5)

# Annotate threshold lines for cluster separation
plt.axhline(y=5, color='green', linestyle='--', label='Threshold for 2 Clusters')
plt.axhline(y=15, color='orange', linestyle='--', label='Threshold for 1 Cluster')

# Step 4: Label the plot
plt.title('Custom Dendrogram with Thresholds')
plt.xlabel('Index of Points')
plt.ylabel('Distance')
plt.legend()

# Show plot
plt.show()

```
