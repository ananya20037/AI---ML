```python
# Question 2: Feature Scaling
# Description: Demonstrate the importance of feature scaling using a K-Means clustering example.

import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate a synthetic dataset with different scales
data = np.array([[1, 2], [2, 3], [3, 3], [8, 8], [9, 9], [10, 10], [100, 200], [150, 250], [200, 300]])

# Apply K-Means clustering without scaling
kmeans = KMeans(n_clusters=3, random_state=42)
labels_no_scaling = kmeans.fit_predict(data)

# Apply feature scaling (Standardization) to the dataset
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply K-Means clustering after scaling
kmeans_scaled = KMeans(n_clusters=3, random_state=42)
labels_scaled = kmeans_scaled.fit_predict(data_scaled)

# Plot the results before scaling
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=labels_no_scaling, cmap='rainbow')
plt.title("K-Means without Scaling")
plt.xlabel("X")
plt.ylabel("Y")

# Plot the results after scaling
plt.subplot(1, 2, 2)
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=labels_scaled, cmap='rainbow')
plt.title("K-Means with Scaling")
plt.xlabel("Scaled X")
plt.ylabel("Scaled Y")

plt.tight_layout()
plt.show()

# Print cluster centers
print("Cluster centers without scaling:", kmeans.cluster_centers_)
print("Cluster centers with scaling:", kmeans_scaled.cluster_centers_)
```
