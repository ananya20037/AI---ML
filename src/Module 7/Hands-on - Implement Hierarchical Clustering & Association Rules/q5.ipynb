```python
# Question 5: Modify a hierarchical clustering algorithm to perform clustering using a custom distance metric on 2D points.
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

# Custom distance function (e.g., Manhattan distance)
def custom_distance(x, y):
    return np.abs(x[0] - y[0]) + np.abs(x[1] - y[1])

# Sample 2D points
data = np.array([[1, 2], [2, 3], [3, 3], [6, 5], [7, 8], [8, 8]])

# Compute the pairwise distance matrix using the custom distance function
dist_matrix = pdist(data, metric=custom_distance)

# Convert distance matrix to a square form
dist_matrix_square = squareform(dist_matrix)

# Perform hierarchical clustering using the custom distance matrix
Z = linkage(dist_matrix, method='single')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.show()
```
