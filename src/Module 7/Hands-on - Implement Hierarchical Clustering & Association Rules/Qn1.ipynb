```python

# Title: Perform Hierarchical Clustering

# Task 1: Hierarchical Clustering with Simple 2D Data
# Step 1: Import the necessary libraries.
# Step 2: Generate and visualize a simple 2D dataset.
# Step 3: Perform Hierarchical Clustering and plot the dendrogram.

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Step 2: Generate a simple 2D dataset
data = np.array([[1, 2], [2, 3], [3, 3], [6, 5], [7, 8], [8, 8]])

# Step 3: Perform Hierarchical Clustering using the 'single' method
Z = linkage(data, method='single')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()


# Task 2: Hierarchical Clustering with Iris Dataset
# Step 1: Import the Iris dataset.
# Step 2: Perform Hierarchical Clustering and visualize using a dendrogram.
# Step 3: Determine the number of clusters from the dendrogram and interpret.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram

# Step 1: Import the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Perform Hierarchical Clustering using the 'ward' method
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram (Iris Dataset)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Step 3: Determine the number of clusters from the dendrogram
# Visually, we can choose the cut-off distance based on where the clusters are well-separated
# Typically, cutting the dendrogram where the vertical line is the longest would indicate the number of clusters.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# Step 1: Import the Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Perform Hierarchical Clustering using the 'ward' method
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram (Iris Dataset)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Step 3: Determine the number of clusters from the dendrogram
# Let's cut the dendrogram at a specific distance threshold to get the clusters
# Here, we choose a threshold where the vertical distance between clusters is large
threshold = 10  # Set the threshold based on visual inspection of the dendrogram

# Using fcluster to assign clusters based on the threshold
clusters = fcluster(Z, threshold, criterion='distance')

# Print the resulting clusters for each sample
print("Cluster assignments:", clusters)



# Task 3: Hierarchical Clustering with Random Data
# Step 1: Generate random data.
# Step 2: Perform Hierarchical Clustering.
# Step 3: Experiment with different linkage methods such as 'single', 'complete', and observe the differences.
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Step 1: Generate random data
np.random.seed(42)  # For reproducibility
data = np.random.rand(10, 2)  # 10 random 2D data points

# Step 2: Perform Hierarchical Clustering with different linkage methods
linkage_methods = ['single', 'complete', 'ward']

plt.figure(figsize=(12, 8))

for i, method in enumerate(linkage_methods, 1):
    plt.subplot(2, 2, i)
    Z = linkage(data, method=method)
    dendrogram(Z)
    plt.title(f'Hierarchical Clustering Dendrogram ({method} linkage)')

plt.tight_layout()
plt.show()
```
